```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
unique_players = purchase_data['SN'].nunique()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = purchase_data['Item Name'].nunique()
avg_price = purchase_data['Price'].mean()
avg_age = purchase_data['Age'].mean()
total_items = purchase_data['Purchase ID'].nunique()
```

```python
summary_table = pd.DataFrame({'Total Items':[total_items],
                              'Unique Items':items,
                              'Average Price':avg_price,
                              'Average Age':avg_age})


summary_table['Average Age'] = summary_table['Average Age'].map('{:.2f}'.format)
summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)

summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data.groupby(['Gender'])

gender_count = pd.DataFrame(gender_group['SN'].nunique())

gender_dem = gender_count.rename(columns={'SN':'Total Purchasers'})

gender_dem
```

```python
gender_dem['Percent of Total'] = (gender_dem['Total Purchasers']/unique_players)*100

gender_dem['Percent of Total'] = gender_dem['Percent of Total'].map("{:.2f}%".format)

gender_dem
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_count['Average Purchase Price'] = gender_group['Price'].mean()
gender_count['Total Purchase Value'] = gender_group['Price'].sum()
gender_count['Unique Purchasers'] = gender_group['SN'].nunique()
gender_count['Avg Total Purchase per Person'] = (gender_count['Total Purchase Value']/gender_count['Unique Purchasers'])
del gender_count['Unique Purchasers']

gender_count
```

```python
gender_count['Average Purchase Price'] = gender_count['Average Purchase Price'].map("${:.2f}".format)
gender_count['Avg Total Purchase per Person'] = gender_count['Avg Total Purchase per Person'].map("${:.2f}".format)
gender_count['Total Purchase Value'] = gender_count['Total Purchase Value'].map("${:.2f}".format)

gender_count
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,1000]

age_bins=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data['Age Groups'] = pd.cut(purchase_data['Age'], bins, labels=age_bins)

age_group = purchase_data.groupby(['Age Groups'])
ag_pd = pd.DataFrame(age_group['SN'].nunique())

ag_pd = ag_pd.rename(columns={'SN':'Total Purchasers'})
```

```python
ag_pd['Percentage of Players'] = (ag_pd['Total Purchasers']/unique_players)*100

ag_pd['Percentage of Players'] = ag_pd['Percentage of Players'].map("{:.2f}%".format)

ag_pd
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase = pd.DataFrame(age_group['Purchase ID'].count())
age_purchase

age_purchase['Avg. Purchase Price'] = age_group['Price'].sum()/age_purchase['Purchase ID']
age_purchase['Unique Purchasers'] = age_group['SN'].nunique()
age_purchase['Avg. Purchase Total/Person'] = age_group['Price'].sum()/(age_purchase['Unique Purchasers'])
age_purchase['Total Spent'] = age_group['Price'].sum()

del age_purchase['Unique Purchasers']
```

```python
age_purchase['Avg. Purchase Price'] = age_purchase['Avg. Purchase Price'].map("${:.2f}".format)
age_purchase['Avg. Purchase Total/Person'] = age_purchase['Avg. Purchase Total/Person'].map("${:.2f}".format)
age_purchase['Total Spent'] = age_purchase['Total Spent'].map("${:.2f}".format)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_group = purchase_data.groupby(['SN'])
sn_df = pd.DataFrame(sn_group['Purchase ID'].count())

sn_df['Total Purchased'] = sn_group['Price'].sum()
sn_df['Avg Purchase Price'] = sn_df['Total Purchased']/sn_df['Purchase ID']
```

```python
sn_df = sn_df.sort_values(by=['Total Purchased'],ascending=False)
sn_df['Total Purchased'] = sn_df['Total Purchased'].map("${:.2f}".format)
sn_df['Avg Purchase Price']= sn_df['Avg Purchase Price'].map("${:.2f}".format)
sn_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = purchase_data.groupby(['Item ID','Item Name'])
item_df = pd.DataFrame(item_group['Purchase ID'].count())
item_df['Total Purchase Value'] = item_group['Price'].sum()
item_df['Item Price'] = item_df['Total Purchase Value']/item_df['Purchase ID']

item_df = item_df.sort_values(by='Purchase ID',ascending=False)
```

```python
item_df['Item Price'].map("${:.2f}".format)
item_df['Total Purchase Value'].map("${:.2f}".format)

item_df.head()
```
